```python
from roboflow import Roboflow
import os
from roboflow import Roboflow
import os
import concurrent.futures
from roboflow import Roboflow
from tqdm import tqdm

# Initialize the Roboflow object with your private API key
rf = Roboflow(api_key="RrtQrCYClNiblb09ILgk")

# Specify the workspace and project
workspace_name = 'gun-detection-4-deld0'
project_id = 'gun-detection-4'

# Access the project in the specified workspace
project = rf.workspace(workspace_name).project(project_id)
```

```python
import glob

images_folder = []
images_folder.extend(glob.glob("/home/research/Kilsar_Sentinal/Data1/images"))
images_folder.extend(glob.glob("/home/research/Kilsar_Sentinal/Data1/*/images"))
images_folder.extend(glob.glob("/home/research/Kilsar_Sentinal/Data1/*/*/images"))
images_folder.extend(glob.glob("/home/research/Kilsar_Sentinal/Data1/*/*/*/images"))
images_folder.extend(glob.glob("/home/research/Kilsar_Sentinal/Data1/*/*/*/*/images"))
images_folder.sort()
len(images_folder)
```

```python
# Function to find the label folder dynamically
def get_label_folder(image_folder):
    # Navigate one step back from the image folder and append 'labels'
    base_folder = os.path.dirname(image_folder)  # Go one step back
    label_folder = os.path.join(base_folder, 'labels')  # Access 'labels' folder
    return label_folder

# Function to upload a single image with its label
def upload_image_with_label(image_path):
    try:
        # Find the corresponding label for the image
        image_name = os.path.basename(image_path)
        label_path = os.path.join(get_label_folder(os.path.dirname(image_path)), os.path.splitext(image_name)[0] + '.txt')  # Assuming labels are in .txt

        # Check if the label file exists
        if os.path.exists(label_path):
            project.upload(
                image_path=image_path,
                annotation_path=label_path,  # Upload the label
                batch_name="gun_annotated_batch",  # Customize as needed
                num_retry_uploads=3
            )
        else:
            print(f"Label for {label_path} not found.")
            project.upload(
                image_path=image_path,
                batch_name="gun_unannotated_batch",  # Customize as needed
                num_retry_uploads=3
            )
    except Exception as e:
        print(f"Failed to upload {image_path}: {e}")
    
# Function to process images using multiprocessing
def process_image(image_folder):
    # Collect all image paths
    images = [os.path.join(image_folder, img) for img in os.listdir(image_folder) if img.endswith('.jpg')]

    # Multiprocessing to upload images in parallel
    with concurrent.futures.ProcessPoolExecutor() as executor:
        executor.map(upload_image_with_label, images)



# Collect all image paths
for image_folder in images_folder:
    process_image(image_folder)

    print("Upload process completed.", images_folder)
```
