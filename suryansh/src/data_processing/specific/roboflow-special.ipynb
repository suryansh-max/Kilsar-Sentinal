```python
import os
import concurrent.futures

# Define the class mapping
class_mapping = {
    0: 'gun',
    1: 'gun'
}

# Directory containing the label files
label_directory = '/home/research/Kilsar_Sentinal/Data1/Threat Detection/train/labels'

def process_file(filename):
    file_path = os.path.join(label_directory, filename)
    
    # Read the file
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Process the file
    new_lines = []
    for line in lines:
        parts = line.strip().split()
        if len(parts) > 1:
            class_id = int(parts[0])
            if class_id in class_mapping:
                # Replace class ID with the class name
                new_class_name = class_mapping[class_id]
                new_line = f"{new_class_name} {' '.join(parts[1:])}\n"
                new_lines.append(new_line)
            elif class_id == 2:
                # Skip lines with class ID 2
                continue
        else:
            # Handle unexpected line formats if needed
            continue
    
    # Write the modified file
    with open(file_path, 'w') as file:
        file.writelines(new_lines)

def main():
    # Get the list of all label files
    label_files = [f for f in os.listdir(label_directory) if f.endswith('.txt')]
    
    # Use ProcessPoolExecutor to process files in parallel
    with concurrent.futures.ProcessPoolExecutor(max_workers=50) as executor:
        executor.map(process_file, label_files)


main()
```
